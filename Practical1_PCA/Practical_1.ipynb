```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
data = pd.read_csv("E:\\Machine Learning Practicals\\datasets\\Wine.csv")
data.head(5)
```

```python
data.info()
```

```python
# Standardization
scalar = StandardScaler()
scalar.fit(data)
```

```python
scaled_data = scalar.transform(data)
scaled_data
# Mean 0, SD 1. All values will be ranging from -3 to +3
```

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:, 0,], data_pca[:, 1], c=data['Customer_Segment'], cmap='plasma')
plt.xlabel("First Principal Camponent")
plt.ylabel("Second Principal Camponent")
```
